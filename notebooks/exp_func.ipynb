This notebook generates dummy exponential weighting functions used in data preprocessing, i.e., creating the spatial terms in the linear causal inference model.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
window_size = 25
```

```python
dist_matrix = np.sqrt(
        np.arange(-window_size, window_size + 1)[np.newaxis, :] ** 2
        + np.arange(-window_size, window_size + 1)[:, np.newaxis] ** 2
    )
```

```python
def create_weight_matrix(dist_matrix, window_size, length_scale):
    """This function generates the weight matrix based on the given length scale."""
    weight_matrix = np.exp(-dist_matrix / length_scale)
    weight_matrix[window_size, window_size] = 0
    weight_matrix /= weight_matrix.sum()
    return weight_matrix
```

```python
ndvi_w = create_weight_matrix(dist_matrix, window_size, 15)
plt.imshow(ndvi_w, cmap='viridis')
plt.savefig('../figures/exp_func.png', dpi=300, bbox_inches='tight')
```

```python
ndvi_w = create_weight_matrix(dist_matrix, window_size, 15)
plt.imshow(ndvi_w, cmap='plasma')
plt.savefig('../figures/exp_func2.png', dpi=300, bbox_inches='tight')
```
