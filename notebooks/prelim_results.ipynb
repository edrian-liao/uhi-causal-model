# Studying Effects of Certain Parameters to Ridge R2 Scores (Boston)

This notebook shows the impact of changing some parameters to the R2 scores. This helps in figuring out the optimla range for each parameter during cross-validation.

````python
import matplotlib.pyplot as plt
import csv
import numpy as np
import pandas as pd
import os
````

````python
base_path = "../results/joint/"
````

## Truncated NDVI values

````python
df1 = pd.read_csv(os.path.join(base_path, "ndvi/ndvi_unchanged.csv"))
df2 = pd.read_csv(os.path.join(base_path, "ndvi/ndvi_changed.csv"))
````

````python
ridge = np.array([df1['ridge_r2'], df2['ridge_r2']])
gp = np.array([df1['gp_r2'], df2['gp_r2']])
total = np.array([df1['total_r2'], df2['total_r2']])
````

````python
iter = range(0,20)
plt.figure(figsize=(10,5))
plt.plot(iter, ridge[0], label="With (-) NDVI values", color="blue")
plt.plot(iter, ridge[1], label="Without (-) NDVI values", color="orange")
plt.xlabel("Iterations")
plt.ylabel("Ridge R2 Score")
plt.title("Ridge R2 Scores with/out Negative NDVI Values")
plt.legend()
plt.grid(True)
````

````python
iter = range(0,20)
plt.figure(figsize=(10,5))
plt.plot(iter, gp[0], label="With (-) NDVI values", color="blue")
plt.plot(iter, gp[1], label="Without (-) NDVI values", color="orange")
plt.xlabel("Iterations")
plt.ylabel("GP R2 Score")
plt.title("GP R2 Scores with/out Negative NDVI Values")
plt.legend()
plt.grid(True)
````

````python
iter = range(0,20)
plt.figure(figsize=(10,5))
plt.plot(iter, total[0], label="With (-) NDVI values", color="blue")
plt.plot(iter, [np.mean(total[0])]*20, label = "With (-) NDVI values (Mean)", linestyle="dashed", color="blue")
plt.plot(iter, total[1], label="Without (-) NDVI values", color="orange")
plt.plot(iter, [np.mean(total[1])]*20, label = "Without (-) NDVI values (Mean)", linestyle="dashed", color="orange")
plt.xlabel("Iterations")
plt.ylabel("Total R2 Score")
plt.title("Total R2 Scores with/out Negative NDVI Values")
plt.legend()
plt.grid(True)
````

Not much impact is seen on the ridge R2 scores when using negative NDVI values or not. However, GP fits better without negative NDVI values, i.e., truncated negative values to 0.

## Noise

````python
df1 = pd.read_csv(os.path.join(base_path, "noise/noise_1e-4.csv"))
df2 = pd.read_csv(os.path.join(base_path, "noise/noise_1e-3.csv"))
df3 = pd.read_csv(os.path.join(base_path, "noise/noise_1e-2.csv"))
df4 = pd.read_csv(os.path.join(base_path, "noise/noise_1e-1.csv"))
df5 = pd.read_csv(os.path.join(base_path, "noise/noise_0.5.csv"))
df6 = pd.read_csv(os.path.join(base_path, "noise/noise_1.csv"))
df7 = pd.read_csv(os.path.join(base_path, "noise/noise_2.csv"))
df8 = pd.read_csv(os.path.join(base_path, "noise/noise_4.csv"))
````
